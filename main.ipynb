```python
import pandas as pd

datos_user = "datos_andres"
saltarse = 10

datos_x_cpu = pd.read_csv(f'ACSI/actividad-2/{datos_user}/cpu_stats.csv')
datos_x_cpu
```

```python
# Limpiamos los primeros 5 registros, para que no afecten el análisis
datos_x_cpu = datos_x_cpu[saltarse:]
datos_x_cpu
```

```python
# mean of %CPU (global), %CPU (user), %CPU (system)
mean_cpu_global = datos_x_cpu["%CPU (global)"].mean()
mean_cpu_user = datos_x_cpu["%CPU (user)"].mean()
mean_cpu_system = datos_x_cpu["%CPU(system)"].mean()

# max of %CPU (global), %CPU (user), %CPU (system)
max_cpu_global = datos_x_cpu["%CPU (global)"].max()
max_cpu_user = datos_x_cpu["%CPU (user)"].max()
max_cpu_system = datos_x_cpu["%CPU(system)"].max()

# min of %CPU (global), %CPU (user), %CPU (system)
min_cpu_global = datos_x_cpu["%CPU (global)"].min()
min_cpu_user = datos_x_cpu["%CPU (user)"].min()
min_cpu_system = datos_x_cpu["%CPU(system)"].min()

# std of %CPU (global), %CPU (user), %CPU (system)
std_cpu_global = datos_x_cpu["%CPU (global)"].std()
std_cpu_user = datos_x_cpu["%CPU (user)"].std()
std_cpu_system = datos_x_cpu["%CPU(system)"].std()

print("mean of %CPU (global):", mean_cpu_global)
print("mean of %CPU (user):", mean_cpu_user)
print("mean of %CPU (system):", mean_cpu_system)
print("max of %CPU (global):", max_cpu_global)
print("max of %CPU (user):", max_cpu_user)
print("max of %CPU (system):", max_cpu_system)
print("min of %CPU (global):", min_cpu_global)
print("min of %CPU (user):", min_cpu_user)
print("min of %CPU (system):", min_cpu_system)
print("std of %CPU (global):", std_cpu_global)
print("std of %CPU (user):", std_cpu_user)
```

```python
import matplotlib.pyplot as plt

span = 15

# Ahora en una grafica
plt.title("CPU stats")
plt.xlabel("Tiempo (s)")
plt.ylabel("Porcentaje de uso (%)")
plt.plot(datos_x_cpu["%CPU (global)"].ewm(span=span).mean(), label="%CPU (global)")
plt.plot(datos_x_cpu["%CPU (user)"].ewm(span=span).mean(), label="%CPU (user)")
plt.plot(datos_x_cpu["%CPU(system)"].ewm(span=span).mean(), label="%CPU (system)")
plt.legend()

plt.show()
```

```python
datos_x_memoria = pd.read_csv(f'ACSI/actividad-2/{datos_user}/memory_stats.csv')
datos_x_memoria

# Limpiamos los primeros 5 registros, para que no afecten el análisis
datos_x_memoria = datos_x_memoria[saltarse:]
```

```python
# Ahora en una grafica
plt.title("Memory stats")
plt.xlabel("Tiempo (s)")
plt.ylabel("Capacidad utilizada (MB)")

plt.plot(datos_x_memoria["Capacidad utilizada"].ewm(span=span).mean(), label="Capacidad utilizada")
plt.legend()
plt.show()

plt.title("Memory stats")
plt.xlabel("Tiempo (s)")
plt.ylabel("Porcentaje de uso (%)")
plt.plot(datos_x_memoria["% Memoria utilizada"].ewm(span=span).mean(), label="% Memoria utilizada")
plt.legend()
plt.show()
```
